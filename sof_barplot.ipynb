```python
import csv
from collections import Counter
from matplotlib import pyplot as plt

plt.style.use('fivethirtyeight')
with open('clean_survey_data.csv', 'r') as f:
    csv_reader = csv.DictReader(f)

    first_counter = Counter()
    second_counter = Counter()

    for line in csv_reader:
        first_languages = line['LanguageWorkedWith'].split(';')
        second_languages = line['LanguageDesireNextYear'].split(';')
        first_counter.update(first_languages)
        second_counter.update(second_languages)

present_languages = []
dev_using1 = []
next_languages = []
dev_using2 = []

for key, value in first_counter.most_common(10):
    present_languages.append(key)
    dev_using1.append(value)

for k, v in second_counter.most_common(10):
    next_languages.append(k)
    dev_using2.append(v)

present_languages.reverse()
dev_using1.reverse()
next_languages.reverse()
dev_using2.reverse()

fig1, (ax1) = plt.subplots()
fig2, (ax2) = plt.subplots()
ax1.barh(present_languages, dev_using1)
ax2.barh(next_languages, dev_using2)

ax1.set_title('Language Worked With')
ax1.set_ylabel('Languages')

ax2.set_title('Language Learned on Next Year')
ax2.set_xlabel('Number of Developer')
ax2.set_ylabel('Languages')

plt.tight_layout()

plt.show()
```
